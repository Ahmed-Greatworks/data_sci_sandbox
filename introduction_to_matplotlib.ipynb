## Making figures with Numpy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histgram plot
* Subplots

```python
# Create data and plot line plot
x = np.linspace(0, 10, 10)
y = np.linspace(10, 20, 10)

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, x**2)
plt.show()
```

```python
# Use the same data to make Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
nut_butter = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter.keys(), nut_butter.values())
ax.set(title="Nut Butter Store", ylabel="Prices ($)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter.keys()), list(nut_butter.values()))
ax.set_xlabel("Prices ($)")
plt.show()
```

```python
# Make some data for histgram plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax1.plot(x, x / 2)
ax2.scatter(x, x**2)
ax3.bar(nut_butter.keys(), nut_butter.values())
ax4.barh(list(nut_butter.keys()), list(nut_butter.values()))

plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter.keys(), nut_butter.values())
ax[1, 1].hist(np.random.randn(1000))

plt.show()
```

### Which one should you use? (pyplot method or OOP method)
* When plotting something quickly, it is okay to use pyplot method
* When plotting something more advance/a-picture-what-more-than-a-thousand-word use OOP method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10, 6))
plt.show()
```

```python
# OOP method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)
plt.show()

# ax.set_xlim([45, 100])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax.set(title="Heart disease and Cholesterol levels", xlabel="Age", ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="target")

ax.axhline(over_50["chol"].mean(), linestyle="--")
plt.show()
```

```python
over_50.head(10)
```

```python
# Subplots of age, chol and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
# Customise plot
ax0.set(title="Heart disease and Cholesterol Levels", ylabel="Cholesterol")
# Add legend
ax0.legend(*scatter.legend_elements(), title="target")
# Adding lines
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Second subplot
# Add data
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
# Customise plot
ax1.set(title="Heart disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
# Add legend
ax1.legend(*scatter1.legend_elements(), title="target")
# Adding lines
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```

```python
# Setting style to seaborn-whitegrid
plt.style.use("seaborn-v0_8-whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")

ax.set(title="Heart disease and Cholesterol levels", xlabel="Age", ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="target")

ax.axhline(over_50["chol"].mean(), linestyle="--")
plt.show()
```

```python
# Customising this plot we have before to make it nicer!
# Subplots of age, chol and thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")
# Customise plot
ax0.set(title="Heart disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.set_xlim([50, 80])
# Add legend
ax0.legend(*scatter.legend_elements(), title="target")
# Adding lines
ax0.axhline(y=over_50["chol"].mean(), linestyle="--")

# Second subplot
# Add data
scatter1 = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter")
# Customise plot
ax1.set(title="Heart disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
ax0.set_xlim([50, 80])
ax1.set_ylim([50, 250])
# Add legend
ax1.legend(*scatter1.legend_elements(), title="target")
# Adding lines
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```
